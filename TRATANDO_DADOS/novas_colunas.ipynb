```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/customers.csv", sep=";")
```

```python
df["Points_double"] = df["Points"] * 2
df
```

```python
df.dtypes
```

```python
#df[["Points", "Points_double"]].astype(float)
df[["Points", "Points_double"]] = df[["Points", "Points_double"]].apply(pd.to_numeric, errors="coerce") # Para converter mais de uma coluna de uma vez use o metodo apply com pd.to_numeric

df.dtypes
```

```python
df["Points_ratio"] = df["Points_double"] / df["Points"]
df
```

```python
df["Constante"] = None
df
```

```python
df["Points_log"] = np.log(df["Points"])
df
```

```python
np.log(df[["Points","Points_double","Points_ratio"]])
```

```python
nomes_alta = []
for i in df['Name']:
    nomes_alta.append(i.upper())

df["Nome_Alta"] = nomes_alta
df
```

```python
df["Name"].str.upper()
```

```python
def get_first(nome:str):
    nome = nome.upper()
    return nome.split("_")[0]

get_first("Téo")
```

```python
df["Name_First"] = df["Name"].apply( get_first )
df
```

```python
df["Name"].apply( lambda x: x.upper().split("_")[0] )
```

```python
def intervalo_pontos(pontos):
    if pontos < 2500:
        return "baixo"
    elif pontos < 3500:
        return "medio"
    else:
        return "alto"
    
df["Faixa_Pontos"] = df["Points"].apply(intervalo_pontos)
df
```
